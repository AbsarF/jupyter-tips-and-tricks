```python
num = 150
x = np.linspace(2.5, 300, num)
y = randn(num)
```

```python
plt.scatter(x, y)
```

```python
y > 1
```

```python
y[y > 1]
```

```python
y[(y < 1) | (y > -1)]
```

```python
plt.scatter(x, y, c='b')
plt.scatter(x[(y < 1) & (y > -1)], y[(y < 1) & (y > -1)], c='r')
```

```python
y[~((y < 1) & (y > -1))] = 1.0
plt.scatter(x, y, c='b')
plt.scatter(x, np.clip(y, -0.5, 0.5), color='red')
```
